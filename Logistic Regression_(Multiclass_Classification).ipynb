```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
%matplotlib inline
```

```python
digits = load_digits()
```

```python
dir(digits)
```

```python
digits.data[0]
```

```python
plt.gray()
for i in range(6):
    plt.matshow(digits.images[i])
```

```python
from sklearn.model_selection import train_test_split
```

```python
 X_train, X_test, y_train, y_test = train_test_split(digits.data,digits.target, test_size = 0.2)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
model.fit( X_train,y_train)
```

```python
model.score(X_test, y_test)
```

```python
plt.matshow(digits.images[29])
```

```python
digits.target[29]
```

```python
model.predict([digits.data[29]])
```

```python
model.predict(digits.data[0:5])
```

```python
y_predicts = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm = confusion_matrix(y_test, y_predicts)
cm
```

```python
import seaborn as sn
plt.figure(figsize = (10,7))
sn.heatmap(cm, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

```python
from sklearn import linear_model
logr = linear_model.LogisticRegression(max_iter=1000)

# Fit the model to the training data
logr.fit(X_train, y_train)

# Get the probabilities of the training set
train_probabilities = logr.predict_proba(X_train)

train_probabilities
```

```python
test_probabilities = logr.predict_proba(X_test)
```
