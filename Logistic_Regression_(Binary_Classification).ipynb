```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model 
```

```python
df = pd.read_csv("diabetes.csv")
df
```

```python
plt.scatter(df.Glucose, df.Age, marker = '+', color = "red")
```

```python
from sklearn.model_selection import train_test_split
```

```python
 X_train, X_test, y_train, y_test = train_test_split(df[['Age']], df.Glucose, test_size = 0.1)
```

```python
X_test
```

```python
model = linear_model.LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
model.predict(X_test)
```

```python
model.score(X_test, y_test) # score decribe about the accuracy of the model
```

```python
# probsbility of the x test
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import make_classification
clf = LogisticRegression()
clf.fit(X_train, y_train)

# Predict probabilities
proba = clf.predict_proba(X_test)
proba
```
